```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df =pd.DataFrame(pd.read_csv('C:\\Users\\yarah\\Downloads\\titanic\\test.csv'))
df.head(30)
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df.drop(columns=['Cabin'], inplace=True) 
df.drop_duplicates(inplace=True)
```

```python
#summary statistiacl
df.describe()
```

```python
#plot the disribution of Survived
plt.figure(figsize=(5,5))
sns.countplot (x="Survived",data=df, palette='Blues')
plt.title("Survived distribution")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Pclass', palette='Blues')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.title('Passenger Class Distribution')
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.countplot(x='Sex',data=df,palette=sns.color_palette())
plt.title("Passenger sex Distribution")
plt.show()
```

```python
ages=df['Age'].quantile([0.25, 0.5, 0.75])
ages
```

```python
data=df.groupby(['Sex','Survived']).size().unstack()
data.plot(kind='bar')
plt.xticks(rotation=0)
font2 = {'family':'serif','color':'darkred','size':15}
plt.xlabel('sex')
plt.title("sex disribution with respect to Survived ")
plt.show()
```

```python
data=df.groupby(['Pclass','Survived']).size().unstack()
data.plot(kind='bar')
plt.xticks(rotation=0)
plt.title("sex disribution with respect to Survived ")
plt.show()
```

```python
df['age_groups']=pd.cut(df['Age'],bins=[0,18,30,60,100],labels=['teenager','young adult','Adult','old persons'])
survival_age=df.groupby('age_groups')['Survived'].mean()
x=survival_age.index
y=survival_age.values
print(x)
print(y)
```

```python
plt.bar(x,y,color={'b','r','y','c'},width=0.5)
plt.xlabel('age_groups')
plt.ylabel('survival_age')
plt.title("Age disribution with respect to Survived")
plt.show()
```
